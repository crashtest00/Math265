```python
#Example 3

#2D plot Template

fig = plt.figure()

#set the domain for t
t = np.arange(-10,10,.1)

#2D Formatting
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect('equal')
axes.set_ylim(bottom=-20,top=20)
axes.set_xlim(left=-20,right=20)
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)

x1 = .2*t
y1 = .3*t

x2 = np.cos(t)
y2 = np.sin(t)

x_sum= x1+x2 
y_sum= y1+y2

x_scale = 5*x_sum
y_scale = 5*y_sum


#Graphs go here
#axes.plot(x1,y1,'r')
#axes.plot(x2,y2,'b')
axes.plot(x_sum, y_sum, 'b')
axes.plot(x_scale, y_scale, 'r')
```

```python
#Example 4 Adding and scaling vector–valued functions.
#A cycloid is a graph traced by a point p on a rolling circle, as shown in Figure 12.6.
#Find an equation describing the cycloid, where the circle has radius 1.

#2D plot
fig = plt.figure()


#2D Formatting
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect('equal')
axes.set_ylim(bottom=0,top=15)
axes.set_xlim(left=0,right=15)
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)

#set the domain for t
t = np.arange(0,15,.1)

#circle
x1 = np.cos(t)
y1 = -np.sin(t)

#translation
x2= 1*t
y2= 1

#add the functions together
x_sum = x1+x2
y_sum = y1+y2

#Graphs go here
#axes.plot(x1,y1,'r')
#axes.plot(x2,y2,'b')
axes.plot(x_sum, y_sum, 'b')
```

```python
#Example 5 Finding and graphing displacement vectors
#Let r(t) =⟨cos(π/2t), sin(π/2t)⟩. Graph r(t) on -1<=t<=1, and find the displacement
#of r(t) on this interval.

#2D plot
fig = plt.figure()


#2D Formatting
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect('equal')
axes.set_ylim(bottom=-1,top=1)
axes.set_xlim(left=-1,right=1)
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)

#set the domain for t
max=1
min=-1
t = np.arange(min,max,.1)

def getx(v):
    return np.cos(v*np.pi/2)
   
def gety(v):
    return np.sin(v*np.pi/2)

#displacement
x_disp= (getx(max)-getx(min))*t #Change in x with respect to t
y_disp= (gety(max)-gety(min))*t #Change in y with respect to t

#graph
axes.plot(x_disp,y_disp,'r')
axes.plot(getx(t),gety(t),'b')
```

```python
# Example 6 Average rate of change
# Let r(t) =⟨cos(π/2t); sin(π/2t)⟩ as in Example 5. Find the average rate of change of r(t) on [-1; 1] and on [-1; 5].

#2D plot

fig = plt.figure()


#2D Formatting
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect('equal')
axes.set_ylim(bottom=-1,top=1)
axes.set_xlim(left=-1,right=1)
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)

#set the domain for t
max=5 #change this to 5 for the second half of the example
min=-1
t = np.arange(min,max,.1)

def x(val):
    return np.cos(val*np.pi/2) #equation for x value goes here

def y(val):
    return np.sin(val*np.pi/2) #equation for y value goes here

def v(val):
    return np.array([x(val),y(val)]) #This creates a 2D vector pair

v0=v(min)
v1=v(max)
t0_t1 = max - min

avg_rate_of_change = (v1-v0)/(t0_t1)
print 'r0 = ', v0
print 'r1 = ', v1
print 'Average rate of change: ', avg_rate_of_change
```
