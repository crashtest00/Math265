```python
# Example 5 Projectile Motion
# Sydney shoots her Red Ryder® bb gun across level ground from an elevation of
# 4ft, where the barrel of the gun makes a 5◦ angle with the horizontal. Find how
# far the bb travels before landing, assuming the bb is fired at the advertised rate
# of 350ft/s and ignoring air resistance.

g = 32 #Gravity = 9.81m/s^2 or 32ft/s^2
vi = 350 #Initial velocity
xpos = 0 #initial x position
ypos = 4 #initial y position
angle = np.deg2rad(5) #Launch angle in radians 

#Position
def x(time):
    return vi*np.cos(angle)*time+xpos # x function goes here
def y(time):
    return -1/2.0*g*pow(time,2)+vi*np.sin(angle)*time+ypos # y function goes here

#Velocity
def xPrime(time):
    return vi*np.cos(angle) #Derivative of the x function
def yPrime(time):
    return -g*time+vi*np.sin(angle) #Derivative of the y function. 


#Acceleration
def xPrimePrime(time):
    return 0
def yPrimePrime(time):
    return g

#10% Increase
def bump(val):
    return val+val*.1

#Quadratic Formula
def quadratic(a,b,c,):
    plus = (-b+sqrt(pow(b,2)-4*a*c))/(2*a)
    minus = (-b-sqrt(pow(b,2)-4*a*c))/(2*a)
    if minus < 0.0 and plus < 0.0:
        print 'INVALID TIME!!!'
        #minus = 1/0.0 #Elegant crash. Like a tapdancing hippopotamus.
    return plus if plus >= 0.0 else minus
    

# Find time 
def findT():
    return quadratic(-1/2.0*g,vi*np.sin(angle),ypos)   
time = findT()

#Find max altitude (when yPrime = 0)
yMaxTime = vi*np.sin(angle)/g
yMax = y(yMaxTime)

#Find distance traveled
xMax = x(time)

print 'Duration of projectile motion:',time,' seconds' 
print 'Distance traveled: ',xMax,'feet'

#set t variable range 
t = np.arange(0,time,.1)
                    
#2D Formatting
fig = plt.figure(figsize=(10,10))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect(10)
axes.set_ylim(bottom=0,top=bump(yMax))
axes.set_xlim(left=bump(xpos),right=bump(xMax))
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)
                    
#Position
axes.plot(x(t), y(t), 'b')

#Velocity at t1 & t2
#axes.quiver(x(t),y(t),xPrime(t), yPrime(t), scale_units='x', scale=2.5, color = 'r') 
```

```python
# Example 6 Projectile Motion
# Alex holds his sister’s bb gun at a height of 3ft and wants to shoot a target that
# is 6ft above the ground, 25ft away. At what angle should he hold the gun to hit
# his target? (We still assume the muzzle velocity is 350ft/s.)

#10% Increase
def bump(val):
    return val+val*.1

#Quadratic Formula
def quadratic(a,b,c,):
    plus = (-b+sqrt(pow(b,2)-4*a*c))/(2*a)
    minus = (-b-sqrt(pow(b,2)-4*a*c))/(2*a)
    return [plus,minus]

g = 32 #Gravity = 9.81m/s^2 or 32ft/s^2
vi = 350 #Initial velocity
xpos = 0 #initial x position
ypos = 3 #initial y position
#angle = np.deg2rad(5) #Launch angle in radians 
xFinal = 25
yFinal = 6

#Position
def x(time,angle):
    return vi*np.cos(angle)*time+xpos # x function goes here
def y(time,angle):
    return -1/2.0*g*pow(time,2)+vi*np.sin(angle)*time+ypos # y function goes here

#Velocity
def xPrime(time):
    return vi*np.cos(angle) #Derivative of the x function
def yPrime(time):
    return -g*time+vi*np.sin(angle) #Derivative of the y function. 

#Acceleration
def xPrimePrime(time):
    return 0
def yPrimePrime(time):
    return g

#### Find time given position information only
#Solve for cos and sin
# cos() = xFinal/(vi*time)
# sin() = (1/2.0*g*pow(time,2)+yFinal-ypos)/(vi*time)

# Square both equations and set equal to 1 (b/c cos^2 + sin^2 = 1), then multiply by the denominator
# and move everything to one side and set equal to zero:
# pow(xFinal,2) + pow((yFinal-ypos+1/2.0*g*pow(time,2)),2) - pow(vi,2)*pow(time,2)=0 

# Quadratic parts (quadratic in t^2)
at = pow(1/2.0*g,2) #at^4
bt = 2*yFinal*1/2.0*g-2*ypos*1/2.0*g-pow(vi,2) #bt^2
ct = pow(xFinal,2)+(pow(yFinal,2))+pow(ypos,2)-2*yFinal*ypos #c

qList = sqrt(quadratic(at,bt,ct))
for q in qList:
    print 'Possible time: ', q, 'seconds'
    #Find angle
    print 'Angle = ', np.rad2deg(np.arccos(xFinal/(vi*q))),' degrees'
    
#GRAPH

#set time and angle  
t1 = np.linspace(0,qList[0])
t2 = np.linspace(0,qList[1])
a1 = np.arccos(xFinal/(vi*qList[0]))
a2 = np.arccos(xFinal/(vi*qList[1]))

#High Angle
fig = plt.figure(figsize=(10,10))
axes1 = fig.add_subplot(2,1,1)
axes1.grid(which='major',axis='both',)
axes1.axhline(y=0,color='k',lw=1)
axes1.axvline(x=0,color='k',lw=1)
                    
#Position
axes1.plot(x(t1,a1), y(t1,a1), 'b')

#Low Angle
axes2 = fig.add_subplot(2,1,2)
axes2.grid(which='major',axis='both',)
axes2.axhline(y=0,color='k',lw=1)
axes2.axvline(x=0,color='k',lw=1)
                    
#Position
axes2.plot(x(t2,a2), y(t2,a2), 'b')
```

```python
# Example 7 Distance Traveled, Displacement, and Average Speed
# A particle moves in space with position function r(t) = ⟨t; t^2; sin(πt)⟩ on [-2; 2], 
# where t is measured in seconds and distances are in meters. Find:
# 1. The distance traveled by the particle on [-2; 2].
# 2. The displacement of the particle on [-2; 2].
# 3. The particle’s average speed.

#Position
def x(val):
    return val # x function goes here
def y(val):
    return pow(val,2) # y function goes here
def z(val):
    return np.sin(np.pi*val) # z function goes here

#Velocity
def xPrime(val):
    if val == 0:
        return 1
    else:
        return val/val
def yPrime(val):
    return 2*val #Derivative of the y function
def zPrime(val):
    return np.pi*cos(np.pi*val) #Derivative of the z function

#Acceleration
def xPrimePrime(val):
    if val == 0:
        return 0
    else:
        return 0/val
def yPrimePrime(val):
    if val == 0:
        return 2
    else:
        return 2*val/val
def zPrimePrime(val):
    return -1*pow(np.pi,2)*sin(np.pi*val) #Derivative of the zPrime function.

#Speed
def speed(val):
    return sqrt(pow(xPrime(val),2) + pow(yPrime(val),2) + pow(zPrime(val),2))

def distance(val1,val2):
    return integrate.quad(lambda t: sqrt(1+pow(2*t,2)+pow(np.pi*cos(np.pi*t),2)),val1,val2)

def displacement(val1,val2):
    return [(x(val2)-x(val1)),(y(val2)-y(val1)),(z(val2)-z(val1))]

def avgSpeed(val1,val2):
    return 1/(val2-val1)*distance(val1,val2)[0]

print 'Distance = ', distance(-2.0,2.0)
print 'Displacement = ', displacement(-2.0,2.0)
print 'Average Speed = ',avgSpeed(-2.0,2.0)

#set the domain for t
t = np.arange(-2,2,.1)

#3D plot
fig = plt.figure(figsize=(10,10)) #For large graphs, make sure to turn off scroll: Cell>Current Outputs>Toggle Scroll
ax = plt.axes(projection = '3d')
ax.view_init(15,30) #adjust the up/down and the left/right viewing angle

#set p variable to graph continous function 
p = np.arange(0,6,.1)

#3D output
ax.plot(x(p),y(p),z(p),'b')
#ax.quiver(0,0,0,xPrime(t), yPrime(t), zPrime(t), length=1.1, color='r')
#ax.quiver(x(t),y(t),z(t), #Set the origin of the vector
#          xPrime(t), yPrime(t), zPrime(t), #Vector components
#          length=1, color='r') #Make it pretty
display(ax)
```
