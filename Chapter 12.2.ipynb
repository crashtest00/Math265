```python
#Example 1 Finding limits of vector–valued functions
#Let r⟨t⟩ =⟨sin(t)/t, t^2-3t+3, cos(t)⟩: 
#Find limit r(t) as t approaches 0

def x(val):
    return np.sin(val)/val #equation for x value goes here

def y(val):
    return pow(val,2)-3*val+3 #equation for y value goes here

def z(val):
    return np.cos(val) #equation for z value goes here

def v(val):
    return np.array([x(val),y(val),z(val)]) #This creates a 3D vector

t=0

print "The limit of x as t approaches zero: ", x(t) #Can't just plug 0 into this equation b/c it would divide by zero. Use l'Hopital's rule
print "The limit of y as t approaches zero: ", y(t)
print "The limit of z as t approaches zero: ", z(t)
```

```python
#Example 2 Evaluating continuity of vector–valued functions
#Let ⃗r(t) =⟨sin(t)/t, t^2 - 3t + 3, cos(t))⟩: Determine whether ⃗r is continuous at t = 0 and t = 1.

def x(val):
    return np.sin(val)/val #equation for x value goes here

def y(val):
    return pow(val,2)-3*val+3 #equation for y value goes here

def z(val):
    return np.cos(val) #equation for z value goes here

def v(val):
    return np.array([x(val),y(val),z(val)]) #This creates a 3D vector

t = 0 #Change this to evaluate the limit at 1 and 0

print "The limit of x as t approaches ", t , "is: ", x(t) #Can't just plug 0 into this equation b/c it would divide by zero. Use l'Hopital's rule
print "The limit of y as t approaches ", t , "is: ", y(t)
print "The limit of z as t approaches ", t , "is: ", z(t)
print "If the result for the limit of any function is nan (not a number), the function is not continuous at that value."
```

```python
# Example 3 Derivatives of vector-valued functions
# Let r(t) =⟨t^2; t⟩.
# 1. Sketch r(t) and r′(t) on the same axes.
# 2. Compute r′(1) and sketch this vector with its initial point at the origin and at r(1).

def x(val):
    return pow(val,2) # x function goes here

def xPrime(val):
    return 2*val #Derivative of the x function

def y(val):
    return val # y function goes here

def yPrime(val):
    return val/val #Derivative of the y function. Must use val/val to return 1!

def v(val):
    return np.array([x(val),y(val),z(val)]) #This creates a 3D vector

def vPrime(val):
    return np.array([xPrime(val),yPrime(val)])
                    
r = 1 #Change this to evaluate the limit at 1 and 0

#set p variable to graph continous function 
p = np.arange(-10,10,.1)
                    
#2D Formatting
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_aspect('equal')
axes.set_ylim(bottom=-2,top=2)
axes.set_xlim(left=-4,right=4)
axes.grid(which='major',axis='both',)
axes.axhline(y=0,color='k',lw=1)
axes.axvline(x=0,color='k',lw=1)
                    
axes.plot(x(p), y(p), 'b')
axes.plot(xPrime(p), yPrime(p), 'm') #Comment this line out to clean up the graph. This is the derivative.
axes.plot(xPrime(r), yPrime(r), 'r')
axes.quiver(xPrime(r), yPrime(r), scale_units='x', scale=1)
axes.quiver(1,1,xPrime(r), yPrime(r), scale_units='x', scale=1)
```
